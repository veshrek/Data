### 4.2 Détection d'un problème dans les données

```python
import datetime as dt
```

```python
#Crétation d'une colonne contenant le mois de la soutenance
df['mois_soutenance']=df['Date de soutenance'].dt.month
dico_mois={1:'Janv',2:'Fev',3:'Mar',4:'Avr',5:'Mai',6:'Jui',7:'Juil',8:'Aout',9:'Sept',10:'Oct',11:'Nov',12:'Dec'}
df['Mois']=df['mois_soutenance'].map(dico_mois)
liste_mois=['Janv','Fev','Mar','Avr','Mai','Jui','Juil','Aout','Sept','Oct','Nov','Dec']
```

```python
# réparttion des mois de soutenance toute année
sns.catplot(x="Mois",data=df,kind='count',order=liste_mois)
plt.show()
```

```python
# répartition d'une nombre de soutenance par année
sns.catplot(x='Year',data=df,kind='count')
plt.xticks(rotation =90)
plt.show()
```

```python
df_1984_2018=df[(df['Year']>=1984) & (df['Year']<=2018)]
```

```python
# réparttion des mois de soutenance toute année
g=sns.catplot(x="Mois",data=df_1984_2018,kind='count',order=liste_mois)
plt.show()
```

Pratiquement toutes les thèses sont soutenues en Janvier!

```python

df_2005_2018=df[(df['Year']>=2005) & (df['Year']<=2018)]
```

```python
# Distribution des mois par année entre 2005 et 2018
dist_2005_2018=df_2005_2018.groupby(['Year','Mois'])['Mois'].value_counts()
df_dist_2005_2018=pd.DataFrame(dist_2005_2018).reset_index()
total_par_an_2005_2018=df_dist_2005_2018.groupby(['Year'])['count'].sum()
```

```python
sns.set_palette('Paired')
k=0
for m in liste_mois:
    sns.barplot(x='Year',y='count',data=df_dist_2005_2018[df_dist_2005_2018['Mois']==m],color=sns.color_palette('Paired')[k])
    k+=1
    
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(13,5))
sns.barplot(x='Year',y='count',data=df_dist_2005_2018,hue='Mois')
plt.xticks(rotation=90)
plt.show()
```

```python
# Répartition des mois de soutenance par année entre 2005 et 2018
g=sns.catplot(x='Mois',data=df_2005_2018,kind='count',col='Year',col_wrap=5)
plt.show()
```

```python
sns.catplot(x='Mois',y='count',data=df_dist_2005_2018,kind='bar',capsize=0.1,order=liste_mois)
plt.show()
```
