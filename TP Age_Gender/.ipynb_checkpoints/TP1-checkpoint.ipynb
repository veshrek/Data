# Manipulation et prétraitement de données
## 3.Première manipulation d'un jeu de données

```python
# chargement des bibliothèques numpy, pandas, matplotlib.pyplot et seaborn avec les alias habituels
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Après un premier chargement des données et une première visualisation des premières lignes, le séparateur n'est pas la virgule mais le point virgule.

```python
# chargement du jeu de données au format csv dans un DataFrame df.
df=pd.read_csv('age_gender.csv',sep=";")
```

```python
#visualisation des premières lignes
df.head()
```

```python
#Visualisation des types de données
df.info()
```

L'âge et le genre sont des entiers. Il y a 23705 lignes.

```python
# statistiques sur les données
df.describe()
```

```python
df['genre'].value_counts()
```

Les âges vont de 1 an à 116 ans. Le genre est codé par 0 ou 1. 

On affiche maintenant la répartition des âges par tranche de 5 ans.

```python
#histogramme des âges par tranche de 5 ans.
sns.histplot(x='age',data=df,bins=[5*i for i in range(25)])
plt.show()
```

Même chose mais en utilisant un découpage avec la fonction cut.

```python
# création d'une colonne dans le DataFrame contenant l'intervalle contenant l'âge
df['I_age']=pd.cut(df['age'],bins=[5*i for i in range(25)])
df.head()
```

```python
#décompte des différents intervalle
#df_I_age_decompte=df['I_age'].value_counts()

#diagramme en barre de celui-ci
g=sns.catplot(x='I_age',data=df,kind='count')
g.set(xlabel='âge',ylabel='effectif')
plt.xticks(rotation=90)
plt.show()
```

```python
# création d'une fonction changeant le 0 et le 1 en genre A et B
def genre_AB(n):
    if n==0:
        return 'A'
    else :
        return 'B'

# création d'une colonne genreAB dans le DataFrame
df['genreAB']=df['genre'].apply(genre_AB)
df.head()
```

Représentation de la répartion de l'âge et du genre dans deux suplots

```python
fig, ax =plt.subplots(1,2)

plt.gcf().subplots_adjust(wspace=0.5) 

# premier graphique avec la répartition des ages
plt.subplot(1,2,1)
sns.histplot(x='age',data=df,bins=[5*i for i in range(25)])

# deuxième graphique avec la réparttion des genres.
plt.subplot(1,2,2)
sns.histplot(x='genreAB',data=df)


plt.title("Répartion de l'échantillon par classe d'âges et par genre",loc='right')

plt.show()
```

Distribution des âges par genre

```python
sns.histplot(x='age',data=df,hue='genreAB',bins=[5*i for i in range(25)])
plt.show()
```

avec utilisation de la transparence

```python
sns.histplot(x='age',data=df,hue='genreAB',bins=[5*i for i in range(25)],alpha=0.7)
plt.show()
```

Je ne trouve pas que la transparence avec le paramètre alpha donne de la lisibilité. Je vais plutôt utiliser le paramètre element sur 'step'

```python
sns.histplot(x='age',data=df,hue='genreAB',bins=[5*i for i in range(25)],element='step')
plt.show()
```

âge moyen selon le genre

```python
sns.barplot(x='genreAB',y='age',data=df,errorbar='sd',capsize=0.1)
plt.show()
```
